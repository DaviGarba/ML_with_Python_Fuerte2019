## Line plot

2. Create a 1d-array of 30 data points from range 10 to range 40 and use it like the **x** variable. Create the **y** variable by multiply **x * 2**.
 * Plot the data in a line plot

```python
# Creating x and y
x = np.array(range(10,40))
y = x**2
```

```python
# Set the figure size
plt.figure(figsize = (9,9))

# Plotting the data

plt.plot(x,y,color='saddlebrown', linewidth = 2)

# Setting the title
plt.title("Plot of a quadratic function", fontsize = 15, color = 'black')


# Setting the axis names
plt.xlabel('$\mathbf{x}$',fontsize = 15, color = 'black')
plt.ylabel('$\mathbf{x}^{2}$', fontsize = 15, color = 'black')
# show
plt.grid()
```

3. Use the same data you created at point 2 and plot them by visualizing only **dots** 

```python
# Set the figure size
plt.figure(figsize = (9,9))

# Plotting the data

plt.scatter(x,y,color='cyan', linewidth = 2)

# Setting the title
plt.title("Plot of a quadratic function", fontsize = 15, color = 'blue')


# Setting the axis names
plt.xlabel('$\mathbf{x}$',fontsize = 15, color = 'blue')
plt.ylabel('$\mathbf{x}^{2}$', fontsize = 15, color = 'blue')
# show
plt.grid()
```

### Exponential plot

4. Use the data you have created at point 2, this time the **y** variable must be the exponential **exp()** of the **x**.
 * Plot them in a line plot

```python
# Calculate exponential for all elements in the array.

y = np.exp(x)
```

```python
# Set the figure size
plt.figure(figsize = (9,9))

# Plotting the data

plt.plot(x,y,color='darkviolet', linewidth = 2)

# Setting the title
plt.title("Plot of an exponential function", fontsize = 15, color = 'deeppink')


# Setting the axis names
plt.xlabel('$\mathbf{x}$',fontsize = 15, color = 'deeppink')
plt.ylabel('$e^{x}$', fontsize = 15, color = 'deeppink')
# show
plt.grid()
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = x
z = x**2


# Creates two subplots and unpacks the output array immediately
f, (ax1, ax2) = plt.subplots(1,2, sharey=False)
ax1.plot(x,y, color = 'saddlebrown')
ax1.set_title('Line', color='gold',fontsize = 15)
ax1.set_xlabel('x', color = 'gold')
ax1.set_ylabel('$x$, $x^2$', color = 'black')
ax2.plot(x, z, color = 'saddlebrown')
ax2.set_title('Quadratic Function', color = 'gold', fontsize = 15)
ax2.set_xlabel('x', color = 'gold')
ax1.grid()
ax2.grid()
```
