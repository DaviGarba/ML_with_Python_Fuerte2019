## Basic Statistic

1. Import the basic libraries

```python
# standard Imports
import numpy as np
import pandas as pd
```

#### From now use the following DF

```python
#Let's create a dataframe to work with
array = np.array([[9,7,np.nan,5],
                  [np.nan,6,7,np.nan],
                  [5,15,np.nan,18],
                  [3,12,16,8]])

df1 = pd.DataFrame(data = array, index=['A','B','C','D'],columns = ['Col1','Col2','Col3','Col4'])

#Show
df1
```

**Remember:** Every method performe the computation on the columns (axis=0) 

2. Sum all columns of the df1

```python
# Sum cols
df1.sum(axis=0)
```

3. Sum all rows of df1

```python
# Sum rows
df1.sum(axis = 1)
```

4. Compute the mean of the columns


```python
# mean
df1.mean(axis=0)
```

5. Find the Min and Max value on every columns

```python
# Min on columns
df1.min(axis = 0)
```

```python
# Max on columns
df1.max(axis = 0)
```

```python
# Min on rows
df1.min(axis = 1)
```

```python
# Max on rows
df1.max(axis = 1)
```

6. Grab the index of df1 where there's the max value

```python
# Index max
df1.fillna(0, inplace = True)
max_val = np.where(df1[max(df1)].values==np.max(df1[max(df1)].values))[0][0]
list(df1[max(df1)].index)[max_val]
```

## Correlation 

### Analyze the correlation on the (reg_movie_ratings.csv) 

13. Read the dataset

```python
ratings_df = pd.read_csv('../../Data/reg_movie_ratings.csv')
ratings_df

```

14. Compute the correlation of the target variable (revenue) with the other variables and sort them.

```python
import seaborn as sn
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
corr_revenue = ratings_df.corr()['Revenue']
corr_revenue.sort_values()
```

15. Explore the correlation by the Pairplot method from seaborn. Reshape the df if necessary.

```python
# visualization imports
plt.figure(figsize=(9,9))
sn.pairplot(ratings_df, diag_kind="kde")
```

## Correlation Matrix

16. Create a correlation matrix from the **reg_movie_ratings.csv**

17. Plot the correlation matrix using the heatmap method from seaborn.
