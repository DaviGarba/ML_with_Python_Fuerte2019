## Drop Entry

1. Import the basic libraries to work.

```python
# Basic imports
import pandas as pd
import numpy as np
```

2. Read the file "premier_league.xlsx". **Hint**  "../Data/premier_league.xlsx"

```python
# Open df
premier_df = pd.read_excel('../../Data/premier_league.xlsx')
premier_df
```

#### Use the drop() method

3. Drop the row where Team = Chelsea (not in-place)

```python
# Drop row
team = premier_df['Team']
pos = list(np.where(team.values=='Chelsea')[0])
premier_df_drop_Chelsea = premier_df.drop(pos, axis =0, inplace= False)
```

```python
premier_df_drop_Chelsea
```

4. Drop the row where Team = Everton (not in-place)

```python
team = premier_df['Team']
pos = list(np.where(team.values=='Everton')[0])
premier_df_drop_Everton = premier_df.drop(pos, axis =0, inplace= False)
```

```python
premier_df_drop_Everton
```

5. Drop the rows from index 5 to index 8.

```python
# Drop rows

premier_df_drop_from_5_to_8 = premier_df.drop(range(5,9), axis =0, inplace= False)
```

```python
premier_df_drop_from_5_to_8
```

6. Drop the column 'Pld' (not in-place)

```python
# Drop column
premier_df_drop_Pld = premier_df.drop('Pld', axis =1, inplace= False)
```

```python
premier_df_drop_Pld
```

7. Drop the column 'Pos' 'W' and 'L' at the same time, not in-place.

```python
# Drop columns
premier_df_drop_Pos_W_L = premier_df.drop(['Pos','W','L'], axis =1, inplace= False)
```

```python
premier_df_drop_Pos_W_L
```
